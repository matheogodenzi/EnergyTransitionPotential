```python
import numpy as np
X_latent = np.genfromtxt("../../data/cleaned_data/X_encoded_simple_best.csv", delimiter=",")

import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Perform hierarchical/agglomerative clustering
Z = linkage(X_latent, method='ward')  # 'ward' minimizes the variance of clusters being merged
```

```python
# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, truncate_mode='level', p=2)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd


data = pd.read_csv("../../data/cleaned_data/calculable_covariates.csv")
# Assuming 'data' is your DataFrame or NumPy array
# Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

desired_number_of_components = scaled_data.shape[1]
# Initialize PCA with the number of components you want to keep
pca = PCA(n_components=desired_number_of_components)

# Fit and transform the data
principal_components = pca.fit_transform(scaled_data)

# The 'principal_components' array now contains the transformed data
```

```python
principal_components
```

```python
import numpy as np

# Components (eigenvectors)
components = pca.components_

# Explained variance
explained_variance = pca.explained_variance_

# Loading matrix
loadings = components.T * np.sqrt(explained_variance)
```
