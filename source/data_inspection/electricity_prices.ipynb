# Inspecting electricity prices

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import sklearn as sk
```

```python
df = pd.read_csv("../../data/raw_data/electricity_prices.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df[df[" operatorLabel"]=="Elektrizitätswerke des Kantons Zürich (EKZ)"]
```

```python
operator_prices_df = df.groupby(" operatorLabel")[" total (cts./kWh)"].agg(["mean", "std"]).reset_index()
operator_prices_df.columns = ["operatorLabel", "mean_elec_price (cts/kWh)", "std_elec_price (cts/kWh)"]
```

```python
df_mun = pd.read_csv("../../data/raw_data/electricity_municipality.csv")
df_mun.columns = ["operator", "website", "MunicipalityNumber", "MunicipalityName", "operatorAddress", "operatorPostalCode", "canton"]
df_mun.head()
```

```python
merged_df = pd.merge(df_mun, operator_prices_df, left_on='operator', right_on='operatorLabel')
```

```python
merged_df.head()
```

```python
elec_df = merged_df.drop(["website", "operatorPostalCode", "operatorLabel", "operatorAddress", "canton"], axis=1)
elec_df.head()
```
